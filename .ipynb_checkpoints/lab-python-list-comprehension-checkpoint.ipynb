## 1.

````python
def initialize_inventory(products):
    inventory = {prod:int(input(f"Enter the quantity of {prod}'s available: ")) for prod in products}
    return inventory

#inventory = initialize_inventory(products)
````

## 2.

````python
def get_customer_orders(products):
    amound_order = int(input("How many products would you like to order? "))
    customer_orders = {input(f"Which product would you like to order? Choose from: {products}: ").lower() for _ in range(amound_order)}
    valid_orders = {prod for prod in customer_orders if prod in products}
    return valid_orders

#valid_orders = get_customer_orders(products)
````

# 3.

````python
def calculate_total_price(valid_orders):
    price_products = {prod:float(input(f"Enter the price of {prod}: "))for prod in valid_orders}
    price_order = sum(price_products.values())
    return price_order

#calculate_total_price(valid_orders)
````

## 4.

````python
def updated_inventory(inventory, valid_orders):
    subtracted_inventory = {prod:amound -1 if prod in valid_orders else amound for prod,amound in inventory.items()}
    new_inventory = {prod:amound for prod, amound in subtracted_inventory.items() if amound > 0}
    return new_inventory

def overview_new_inventory (new_inventory):
    print("Updated Inventory:")
    for prod,amound in new_inventory.items():
        print(f"{prod}:{amound}")    

#new_inventory = updated_inventory(inventory, valid_orders)
#overview_new_inventory (new_inventory)
````

# Needs to be somewhere.

````python
def calculate_order_statistics(valid_orders, products):
    total_products_ordered = len(valid_orders)
    percentage_order = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_order

def print_order_statistics (order_statistics):
    total_products_ordered, percentage_order = order_statistics
    print(f"Order statistics:\nTotal Products Ordered: {total_products_ordered}\nPercentage of Products Ordered: {percentage_order}%")
    
#order_statistics = calculate_order_statistics(valid_orders, products)
#print_order_statistics (order_statistics) 
````

# 5.

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = initialize_inventory(products)
valid_orders = get_customer_orders(products)
print()
order_statistics = calculate_order_statistics(valid_orders, products)
print_order_statistics (order_statistics)
print()
updated_inventory(inventory, valid_orders)
price_order = calculate_total_price(valid_orders)
print(f"Total price: {price_order}")
````
